```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
# 1. CONIFIGURATION
start_date = datetime(2025, 1, 1)
num_days = 365 # 1 year of historical data
routes = ['Route_A_Langata', 'Route_B_Kiambu', 'Route_C_Naivasha', 'Route_D_Gilgil', 'Route_E_Naivasha_Rd']
products = {
    'Bread': {'base_demand': 400, 'margin': 50},
    'Mandazi': {'base_demand': 250, 'margin': 40},
    'Cake': {'base_demand': 100, 'margin': 20}
}
```

```python
# 2. HElPER FUNCTIONS
def is_school_term(date):
    # Simulate school terms (roughly 60% of the year, excluding long holidays)
    month = date.month
    return 1 if month not in [4, 8, 12] else 0
```

```python
def is_payday_week(date):
    # Simulate payday spikes (last 3 days of month and first 3 days)
    day = date.day
    return 1 if (day >= 28 or day <= 3) else 0
```

```python
#. 3 DATA GENERATION LOOP
data = []
current_date = start_date

for _ in range(num_days):
    day_name = current_date.strftime('%A')
    is_weekend = 1 if current_date.weekday() >= 5 else 0
    school_term = is_school_term(current_date)
    payday = is_payday_week(current_date)
    
    for route in routes:
         # Route base multiplier (some routes are busier)
        route_multiplier = np.random.uniform(0.8, 1.2) 
        
        for product, specs in products.items():
            base = specs['base_demand'] * route_multiplier
            
            # Apply demand modifiers based on your proposal's factors
            demand = base
            if is_weekend: demand *= 1.25      # Weekends are busier
            if payday: demand *= 1.30          # Payday spending spikes
            if school_term and product in ['Bread', 'Mandazi']: demand *= 1.15 # Kids buy snacks
            
            # Add random noise (real-world unpredictability)
            actual_demand = int(demand + np.random.normal(0, specs['margin']))
            actual_demand = max(10, actual_demand) # Demand can't be negative
            
            # Simulate the bakery's flawed "Yesterday's Production" logic
            # They usually overproduce by 10% or underproduce by 5% randomly
            production_error = np.random.uniform(-0.05, 0.15) 
            dispatched = int(actual_demand * (1 + production_error))
            produced = dispatched # Assume they bake what they dispatch
            
            # Calculate Sales and Returns
            if dispatched >= actual_demand:
                sold = actual_demand
                returned = dispatched - actual_demand
                stockout = 0
            else:
                sold = dispatched
                returned = 0
                stockout = 1 # They ran out of stock!
            
            data.append([
                current_date.strftime('%Y-%m-%d'), route, product, day_name,
                school_term, payday, is_weekend,
                produced, dispatched, sold, returned, stockout
            ])
    current_date += timedelta(days=1)
```

```python
# 4.CREATE AND SAVE DATAFRAME 
columns = [
    'Date', 'Route_ID', 'Product_Type', 'Day_of_Week', 
    'Is_School_Term', 'Is_Payday_Week', 'Is_Weekend', 
    'Produced_Qty', 'Dispatched_Qty', 'Sold_Qty', 'Returned_Qty', 'Stockout_Flag'
]

df = pd.DataFrame(data, columns=columns)
```

```python
# save to CSV
file_name = 'nakuru_bakery_historical_data.csv'
df.to_csv(file_name, index=False)

print(f"Successfully generated {len(df)} rows of data!")
print(f"Saved as '{file_name}'")
print("\nFirst 5 rows:")
print(df.head())
```

## Model Training and Visulaization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
import joblib
```

```python
# load preprocessed data
print("Loading Preprocessed datasets")
X_train = pd.read_csv("X_train.csv")
X_test = pd.read_csv("X_test.csv")
y_train = pd.read_csv("y_train.csv").squeeze() 
y_test = pd.read_csv("y_test.csv").squeeze()
```

```python
print(f"data loaded.. Training samples: {X_train.shape[0]}, Testing samples:{X_test.shape[0]}\n")
```

```python
# train the random Forest model
print("Training Random Forest Regressor")
# n_estimators=100 means 100 decision trees. random_state=42 ensures reproducible results.
model = RandomForestRegressor(n_estimators=100, random_state=42,max_depth=10)
model.fit(X_train, y_train)
print("Model training complete\n")
```

```python
# Make Predictions & Evaluate Accuracy
print("Generating predictions on unseen test data")
y_pred = model.predict(X_test)
```

```python
# calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
rmse =np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
```

```python
print("--- Model Evaluation Metrics ---")
print(f"Mean Absolute Error (MAE): {mae:.2f} units")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f} units")
print(f"R-squared (R²) Score: {r2:.4f} (Closer to 1.0 is better)\n")
```

```python
# Visualization feature importance
print("Analyzing Feature Importance")
# Extract importance scores and map them to column names
features_importances = pd.Series(model.feature_importances_, index=X_train.columns)
```

```python
#Sort from highest to lowest importance
top_features = features_importances.sort_values(ascending=False).head(10)
```

```python
# creating a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=top_features.values, y=top_features.index,palette="viridis")
plt.title('Top 10 Factors Influencing Bakery Demand', fontsize=15, fontweight='bold')
plt.xlabel('Importance Score', fontsize=13)
plt.ylabel('Features', fontsize=13)
plt.tight_layout()
```
